```python
# Data CLeaning and Visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset using pandas which is of titanic disaster list
df= pd.read_csv('train.csv')
```

```python
# head() shows first five rows of the dataset
df.head()
```

```python
# here we use this for datatype and missing value
```

```python
df.info ()
```

```python
# and thi sis used for statistics value of the datasets
df.describe()
```

```python
# we use this for finding out the missing values
df.isnull()
# it shows which values are missing as true and not missing as false
```

```python
df.isnull().sum()
# it counts the missing values and shows the answer
# from which we know that age has 177 missing values and cabin has 682 missing values and embarked has 2 mssing values
```

```python
# now its time fro handling missing values
df['Age']= df["Age"].mean()
```

```python
df.isnull().sum()
# now we can see that the value in ageis not missing
```

```python
df["Cabin"]= df["Cabin"].mode()[0]
df["Embarked"]=df["Embarked"].mode()[0]
```

```python
df.isnull().sum()
```

```python
# to check if there are the values which are duplicates or not
df.duplicated().sum()
```

```python
# Data visualization
# about how many survived and how many didnt survive
# barchart
sns.countplot(x="Survived", data=df)
```

```python
# barcahrt in group
sns.countplot(x="Sex", hue="Survived", data=df)
```

```python
# histogram
sns.histplot(df["Age"], bins=20)
```

```python
# boxplot
sns.boxplot(x="Survived", y="Age", data=df)
```
